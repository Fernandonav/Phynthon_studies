```python
World_population = pd.read_csv('world_population.csv', sep = ",", skiprows = 0)
```

```python
# get 5 samples
World_population.sample(5)
```

```python
#for Rename columns,create  a dictionary 
rename_column = {"CCA3": "Country Code"}
World_population = World_population.rename(columns=rename_column)
World_population
```

```python
# See the type fo the column
World_population.dtypes
```

```python
#get the unique value if a column of the dataframe
World_population["Continent"].unique()
```

```python
#Change the value from continet column
new_labels = ['Asia-AS', 'Europe-EU', 'Africa-FR','Oceania -OC', 'North America-NA',
       'South America-SA']
new_sorter_labels = dict(zip(World_population["Continent"].unique(), new_labels))
World_population = World_population.assign(Continent = World_population["Continent"].map(new_sorter_labels))
World_population.sample(5)
```

```python
# Count how many countries per continent
World_population["Continent"].value_counts()
```

```python
#Create and add a new column
World_population = World_population.assign(World_population_increased = World_population["2022 Population"] - World_population["2020 Population"])
World_population.sample(5)
```

```python
#get a country with max value in World_population_increased
World_population[World_population["World_population_increased"] == World_population["World_population_increased"].max()]
```

```python
 #group by continent and oly show 2 columns
World_population.groupby("Continent")["Continent","World_population_increased"].sum().sort_values("World_population_increased")
```
